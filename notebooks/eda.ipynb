```python
import os
import random
import numpy as numpy
import matplotlib.pyplot as plt 
from PIL import Image
from collections import Counter
```

```python
DATA_DIR="../data/EuroSAT"
CLASSES=sorted(os.listdir(DATA_DIR))
print("Classes found: ",CLASSES)
print("Number of classes: ",len(CLASSES))
```

```python
img_per_class={}
for cls in CLASSES:
    path=os.path.join(DATA_DIR,cls)
    count=len(os.listdir(path))
    img_per_class[cls]=count
    print("Class: ",cls," | Number of Images: ",count)
```

```python
plt.figure(figsize=(12,5))
plt.bar(img_per_class.keys(),img_per_class.values(),color='green', edgecolor='white')
plt.ylabel("Number of images")
plt.title("EuroSAT class distribution")
plt.savefig("../data/class_distribution.png", dpi=150)
plt.show()
```
